# **22.1 ODE Initial Value Problem Statement**
We define a **differential equation** as a relationship between a function independent variable and any of its derivatives. Whereas an **Ordinary differential equation (ODE)** is a differential equation thats derivatives are one dimensional. Is expressed as $\frac{d^nf(x)}{dx^n}$ where $n$ is the order of said differential equation.

A **Partial differential equation (PDE)** is the general form of a differential equation. A $g(x)$ that satisfies a differential equation is known as a **general solution**. A $p(x)$ that satisfies a differential equation and $n$ known values of its derivatives at certain points is known as a **particular solution** for an ODE of $n^{th}$ order.

The **initial value problem** is what we refer to when trying to find a unique solution to an ODE given an initial value.

# **22.2 Reduction of Order**
When dealing with differential equation there are many methods for solving first order equations however it becomes more complex when dealing with diffential equations of higher orders. Oftentimes we must **reduce the order** of our equation to first order to make the solving method simpler.

The **State of system** as a function of time is the collection of ODE relevant dependent variables. We can take the derivate of the state of the system which in turn allows us to reduce the order of the differential equation.

ODE that can be expressed in matrix form are known as **Linear ODEs**


# **22.3 Euler Method**
Let $\frac{dS(t)}{dt}=F(t,S(t))$ be defined as a first order ODE and let $t$ be a numerical grid of the interval $[t_0, t_f]$ with spacing $h$ we assume that $t_0=0$ and $t_f=Nh$ for some positive integer N. Thus, the linear approximation of $S(t)$ around $t_j$ and $t_{j+1}$ can be expressed as $S(t_{j+1})=S(t_j) + hF(t,S(t))$.

This formula is reffered to as the **Explicit Euler Formula** and it allows us to compute an approximation of the state at $S(t_{j+1})$ given the state at $S(t_j)$ in other words it allows us to compute the approximation of a differential equation at different states. This process can be performed in Python as shown below

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
f = lambda t, s: -2 * s  # ODE: dy/dt = -2y
h = 0.1  # Step size
t = np.arange(0, 2 + h, h)  # Numerical grid
s0 = 1  # Initial Condition

# Explicit Euler Method
s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

# Exact solution for comparison
exact = lambda t: np.exp(-2 * t)

plt.figure(figsize=(12, 8))
plt.plot(t, s, 'bo--', label='Approximate (Euler)')
plt.plot(t, exact(t), 'g', label='Exact Solution')
plt.title('Explicit Euler Method vs Exact Solution for dy/dt = -2y')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid()
plt.legend(loc='upper right')
plt.show()
```
